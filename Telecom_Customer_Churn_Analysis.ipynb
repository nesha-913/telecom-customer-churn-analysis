```python
retained_tenure = df[df['Churn'] == 'No']['tenure']
churned_tenure = df[df['Churn'] == 'Yes']['tenure']

statistic, p_value = mannwhitneyu(
    retained_tenure,
    churned_tenure,
    alternative='two-sided'
)

print("Tenure vs Churn")
print("----------------------------")
print(f"Mann-Whitney U Statistic: {statistic:.4f}")
print(f"P-value: {p_value:.10f}")

if p_value < 0.05:
    print("Result: Significant difference in tenure between churned and retained customers.")
else:
    print("Result: No significant difference in tenure between the two groups.")
```

```python
from scipy.stats import chi2_contingency, mannwhitneyu
import pandas as pd

results = []

# Chi-Square tests
categorical_features = [
    'Contract',
    'InternetService',
    'PaymentMethod',
    'TechSupport',
    'OnlineSecurity'
]

for feature in categorical_features:
    table = pd.crosstab(df[feature], df['Churn'])
    chi2, p, dof, expected = chi2_contingency(table)

    results.append({
        'Feature': feature,
        'Test': 'Chi-Square',
        'Statistic': round(chi2, 4),
        'P-value': p,
        'Significant': 'Yes' if p < 0.05 else 'No'
    })

# Mann-Whitney U for Tenure
retained = df[df['Churn'] == 'No']['tenure']
churned = df[df['Churn'] == 'Yes']['tenure']

stat, p = mannwhitneyu(retained, churned, alternative='two-sided')

results.append({
    'Feature': 'tenure',
    'Test': 'Mann-Whitney U',
    'Statistic': round(stat, 4),
    'P-value': p,
    'Significant': 'Yes' if p < 0.05 else 'No'
})

# Mann-Whitney U for Monthly Charges
retained = df[df['Churn'] == 'No']['MonthlyCharges']
churned = df[df['Churn'] == 'Yes']['MonthlyCharges']

stat, p = mannwhitneyu(retained, churned, alternative='two-sided')

results.append({
    'Feature': 'MonthlyCharges',
    'Test': 'Mann-Whitney U',
    'Statistic': round(stat, 4),
    'P-value': p,
    'Significant': 'Yes' if p < 0.05 else 'No'
})

statistical_results = pd.DataFrame(results)

statistical_results
```

```python
def predict_customer_churn(
    gender,
    senior_citizen,
    partner,
    dependents,
    tenure,
    phone_service,
    multiple_lines,
    internet_service,
    online_security,
    online_backup,
    device_protection,
    tech_support,
    streaming_tv,
    streaming_movies,
    contract,
    paperless_billing,
    payment_method,
    monthly_charges,
    total_charges
):
    try:
        # 1. User input එක DataFrame එකකට දානවා
        customer = pd.DataFrame([{
            "gender": gender,
            "SeniorCitizen": int(senior_citizen),
            "Partner": partner,
            "Dependents": dependents,
            "tenure": float(tenure),
            "PhoneService": phone_service,
            "MultipleLines": multiple_lines,
            "InternetService": internet_service,
            "OnlineSecurity": online_security,
            "OnlineBackup": online_backup,
            "DeviceProtection": device_protection,
            "TechSupport": tech_support,
            "StreamingTV": streaming_tv,
            "StreamingMovies": streaming_movies,
            "Contract": contract,
            "PaperlessBilling": paperless_billing,
            "PaymentMethod": payment_method,
            "MonthlyCharges": float(monthly_charges),
            "TotalCharges": float(total_charges)
        }])

        # 2. Categorical values numbers බවට convert කරනවා
        customer_encoded = pd.get_dummies(customer, dtype=int)

        # 3. Training dataset එකේ columns වලට exactly match කරනවා
        customer_encoded = customer_encoded.reindex(
            columns=feature_columns,
            fill_value=0
        )

        # 4. Numerical fields scale කරනවා
        customer_encoded[numerical_columns] = scaler.transform(
            customer_encoded[numerical_columns]
        )

        # 5. Prediction සහ probability ගන්නවා
        prediction = model.predict(customer_encoded)[0]
        churn_probability = model.predict_proba(customer_encoded)[0][1]

        # 6. Risk level එක define කරනවා
        if churn_probability >= 0.70:
            risk_level = "High Risk"
        elif churn_probability >= 0.40:
            risk_level = "Medium Risk"
        else:
            risk_level = "Low Risk"

        prediction_label = (
            "Likely to Churn"
            if prediction == 1
            else "Likely to Stay"
        )

        return (
            prediction_label,
            f"{churn_probability * 100:.2f}%",
            risk_level
        )

    except Exception as error:
        return (
            "Prediction Error",
            "0.00%",
            str(error)
        )
```
